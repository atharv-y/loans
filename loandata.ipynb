```python
catfts = ['purpose']
```

```python
data = pd.get_dummies(data=loans, columns=catfts, drop_first=True, dtype=float)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('not.fully.paid', axis=1), data['not.fully.paid'], test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfcpred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfcpred))
print(classification_report(y_test, rfcpred))
```
